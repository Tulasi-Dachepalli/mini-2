```python
from sklearn.datasets import load_digits

# Load the digits dataset
digits = load_digits()
X = digits.data
y = digits.target

from sklearn.manifold import Isomap

# Create an instance of the Isomap class with 2 components
iso = Isomap(n_components=2)

# Fit the model to the data and transform it to 2D
X_iso = iso.fit_transform(X)

import matplotlib.pyplot as plt

# Create a scatter plot of the projected digits with their true labels as colors
plt.scatter(X_iso[:, 0], X_iso[:, 1], c=y, cmap='viridis')
plt.colorbar()

# Set axis labels and plot title
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('2D Projection of Handwritten Digits using IsoMap')

# Show the plot
plt.show()
```
